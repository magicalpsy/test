```
grad_cam_data = dataCombine[['pre_rsrp', 'pre_sinr', 'pre_Txpwr', 'pre_dl_bler', 'pre_ul_bler', 'pre_dl_mcs', 'pre_dl_rb','pre_ul_mcs']]
```

```
dataCombine.isnull().sum()
```

```
temp_result = []
for index ,temp in enumerate(np.array(grad_cam_data)):
    c = temp.argsort()[-3:][::-1]
    c = c.reshape(1, 3)
    temp_result.append(c)
```

```
c = np.array(temp_result)
a = c.reshape(-1,3)
a = pd.DataFrame(a, columns=['top1', 'top2', 'top3'])
```

```
a['top1'].value_counts()
```

```
df = pd.concat( [dataCombine, a], axis=1 )
```

```
test_file = r'/content/drive/My Drive/WORK/test_last_result.csv'
df.to_csv(test_file , encoding='utf8')
```
